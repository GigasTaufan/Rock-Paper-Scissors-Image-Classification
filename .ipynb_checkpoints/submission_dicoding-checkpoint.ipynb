```python
# validation directory
val_dir = "/tmp/rockpaperscissors/validation/"
val_rock_dir = "/tmp/rockpaperscissors/validation/rock/"
val_paper_dir = "/tmp/rockpaperscissors/validation/paper/"
val_scissors_dir = "/tmp/rockpaperscissors/validation/scissors/"
```

## Split Training dan Validation Data

```python
def split_data(base_dir, train_dir, val_dir, test_size):
    files = []
    for filename in os.listdir(base_dir):
        file = base_dir + filename
        if os.path.getsize(file) > 0:
            files.append(filename)
        else:
            print(filename + " others")

    training_length = int(len(files) * test_size)
    val_length = int(len(files) - training_length)
    shuffled_set = random.sample(files, len(files))
    training_set = shuffled_set[0:training_length]
    val_set = shuffled_set[training_length:]

    for filename in training_set:
        this_file = base_dir + filename
        destination = train_dir + filename
        copyfile(this_file, destination)

    for filename in val_set:
        this_file = base_dir + filename
        destination = val_dir + filename
        copyfile(this_file, destination)
```

```python
# split data rock
split_data(base_rock_dir, train_rock_dir, val_rock_dir, 0.6)
```

```python
# split data paper
split_data(base_paper_dir, train_paper_dir, val_paper_dir, 0.6)
```

```python
# split data scissors
split_data(base_scissors_dir, train_scissors_dir, val_scissors_dir, 0.6)
```

```python
total_train = len(os.listdir(train_rock_dir)) + len(os.listdir(train_paper_dir)) + len(os.listdir(train_scissors_dir))
print(f"Total Train Data: {total_train}")
```

```python
total_val = len(os.listdir(val_rock_dir)) + len(os.listdir(val_paper_dir)) + len(os.listdir(val_scissors_dir))
print(f"Total Validation Data: {total_val}")
```

## Augmentasi Data

```python
train_datagen = ImageDataGenerator(
    rescale = 1./225,
    rotation_range = 20,
    horizontal_flip = True,
    shear_range = 0.2,
    fill_mode = 'nearest'
)
 
test_datagen = ImageDataGenerator(
    rescale = 1./225,
    rotation_range = 20,
    horizontal_flip = True,
    shear_range = 0.2,
    fill_mode = 'nearest'
)
```

```python
train_generator = train_datagen.flow_from_directory(
        train_dir,  # direktori data latih
        target_size=(150, 150),  # mengubah resolusi seluruh gambar menjadi 150x150 piksel
        batch_size=32,
        class_mode='categorical')
 
validation_generator = test_datagen.flow_from_directory(
        val_dir, # direktori data validasi
        target_size=(150, 150), # mengubah resolusi seluruh gambar menjadi 150x150 piksel
        batch_size=32,
        class_mode='categorical')
```

## Membangun Arsitektur CNN

```python
tf.keras.backend.clear_session()

model_class = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(256, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])
```

```python
model_class.summary()
```

```python
model_class.compile(loss = 'categorical_crossentropy',
              optimizer = tf.optimizers.Adam(),
              metrics=['accuracy'])
```

## Melakukan Training Model Dengan Train Data

```python
# train model
model_class.fit(
    train_generator,
    steps_per_epoch = 25,
    epochs = 25,
    validation_data = validation_generator,
    validation_steps = 5,
    verbose =2
)
```

## Evaluasi Model

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
validation_generator = test_datagen.flow_from_directory(
    val_dir,
    batch_size=159,
    class_mode='categorical',
    target_size=(150, 150),
    shuffle=False
)

filenames = validation_generator.filenames
nb_samples = len(filenames)

Y_pred = model_class.predict_generator(validation_generator, steps = nb_samples)
y_pred = np.argmax(Y_pred, axis=1)

print('Confusion Matrix')
print(confusion_matrix(validation_generator.classes, y_pred))

print('Classification Report')
target_names = ['Rock', 'Paper', 'Scissors']
print(classification_report(validation_generator.classes, y_pred, target_names=target_names))
```

Berdasarkan hasil Classification Report menunjukkan akurasi yang sudah cukup bagus untuk model yang dibuat

## Prediksi Gambar Baru

```python
from google.colab import files
from keras.preprocessing import image
import matplotlib.image as mpimg
```

```python
# scissors
uploaded = files.upload()
 
for fn in uploaded.keys():
 
  # predicting images
  path = fn
  img = image.load_img(path, target_size=(150,150))
  imgplot = plt.imshow(img)
  x = image.img_to_array(img)
  x = np.expand_dims(x, axis=0)

  predict_array = model_class.predict(x)[0]

  df = pd.DataFrame(predict_array)
  df = df.rename({0:'Probability'}, axis = 'columns')
  prod = ['Paper', 'Rock', 'Scissors']
  df['Product'] = prod
  df = df[['Product', 'Probability']]
 
  images = np.vstack([x])
  classes = model_class.predict(images, batch_size=10)
  
  print(fn)

  if classes[0,0]!=0:
    print('paper')
  elif classes[0,1]!=0:
    print('rock')
  else:
    print('scissors')

print(df)
```
